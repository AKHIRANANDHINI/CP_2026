```python
class DSU:
    def __init__(self, n):
        self.parent = [i for i in range(n+1)]

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]

    def union(self, a, b):
        pa = self.find(a)
        pb = self.find(b)
        if pa != pb:
            self.parent[pb] = pa


# Main Program
n = int(input("Number of intersections: "))
m = int(input("Number of roads: "))

dsu = DSU(n)

print("Enter roads:")
for _ in range(m):
    a, b = map(int, input().split())
    dsu.union(a, b)

q = int(input("Number of queries: "))
print("Enter queries:")
for _ in range(q):
    a, b = map(int, input().split())
    if dsu.find(a) == dsu.find(b):
        print("YES")
    else:
        print("NO")
```
